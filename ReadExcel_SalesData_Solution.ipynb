This notebook demonstrates the Day 2 tasks:
1. Read sheet "January".
2. Show first 5 rows.
3. Read only columns A to C (which correspond to `date`, `order_id`, `product` in this sample).
4. Ensure `date` is parsed as datetime.
5. Read all sheets and display their names.
6. Count total rows across all sheets.


```
# Import libraries
import pandas as pd
```

```
# 1. Read sheet 'January'
excel_file = '/mnt/data/sales_data.xlsx'
df_jan = pd.read_excel(excel_file, sheet_name='January')
df_jan.head()
```

```
# 2. Show first 5 rows (already shown above via head())
print('First 5 rows of January sheet:')
print(df_jan.head())
```

```
# 3. Read only columns A to C (date, order_id, product) using usecols
cols_df = pd.read_excel(excel_file, sheet_name='January', usecols='A:C', parse_dates=['date'])
print('Columns A:C with date parsed:')
print(cols_df.head())
```

```
# 4. Ensure `date` is datetime and check dtype
print('\nDtypes:')
print(cols_df.dtypes)
```

```
# 5. Read all sheets and display their names
all_sheets = pd.read_excel(excel_file, sheet_name=None)
print('\nSheets found:', list(all_sheets.keys()))
```

```
# 6. Count total rows across all sheets
total_rows = sum(len(df) for df in all_sheets.values())
print('\nTotal rows across all sheets:', total_rows)
```

```
# Extra: Show combined DataFrame
combined = pd.concat(all_sheets.values(), ignore_index=True)
print('\nCombined DataFrame sample:')
print(combined.head())
```
